```python
import torch
import torch.nn as nn
class LoRALayer(nn.Module):
    def __init__(self, in_features, out_features, r, alpha):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features, bias=False)
        for param in self.linear.parameters():
            param.requires_grad = False
        self.A = nn.Parameter(torch.randn(in_features, r) * 0.01)

        self.B = nn.Parameter(torch.zeros(r, out_features))
        self.r = r
        self.scaling = alpha / r
    def forward(self, x):
        original = self.linear(x)

        lora_update = x @ self.A @ self.B
        return original + self.scaling * lora_update

```

```python
layer = LoRALayer(in_features=8, out_features=4, r=2, alpha=4)
x = torch.randn(3, 8)
out = layer(x)
print(out.shape)

total = sum(p.numel() for p in layer.parameters())
trainable = sum(p.numel() for p in layer.parameters() if p.requires_grad)
print(f"Trainable: {trainable} / Total: {total}")
```
